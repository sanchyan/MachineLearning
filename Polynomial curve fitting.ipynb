# Code Implementation of polynomial curve fitting

```python
#import necessary items

import numpy as np
import matplotlib.pyplot as plt
```

```python
#let us take 10 sample points
#variable to draw the sine wave with 100 points
sine_points = 100 

#sample points which we are going to use as training data
sample_points = 10

 # 100 values from 0 to 1
x = np.linspace(0,1, num = sine_points)

#implementation of formula in a variable for convinience
sine = np.sin(2*np.pi*x) 

 #10 values from 0 to 1
x_input = np.linspace(0,1, num = sample_points)


#let us find out the original output as per the formula t = sin(2pix) +eta using gaussian noise formula
t_output = np.sin(2*np.pi*x_input) + 0.2*np.random.normal(0,1,sample_points)

#plot the data
plt.plot(x_input,t_output,'bo',x,sine,'g-')
plt.legend(('Training Data', 'Sine Wave '))
```

```python
#we use figsize to show the custom size of the plot
plt.figure(figsize=(15,10))

#we use j,i in enumerate to iterate the value i and the index will be j
for j,i in enumerate([0,1,3,9]):
    z = np.polyfit(x_input,t_output,i) #np.polyfit is used for fitting polynomials and finding the coefficients
    p = np.poly1d(z) #polynimial construction
    y_output = p(x_input)
    plt.subplot(2,2,j+1)
    plt.plot(x_input,t_output,'bo',x_input,y_output,'r-',x,sine,'g-')
    plt.legend(('Training Data', 'Fitting with M = '+str(i),'Sine Wave '))
    
```

# Error Calculation on the overfitted model



```python
#let us take a sample of polynomial order 9 for 10 points of training data 
M = 9
zn = np.polyfit(x_input,t_output,M)
pn = np.poly1d(zn)
yn = pn(x_input)

plt.plot(x_input,yn,'g-',x_input,t_output,'bo')
plt.legend(('M ='+str(M),'Training Data'))


#error Calculation
error = 0.5*(np.sum((yn-t_output)**2))
print(error)
```

# Let us check for the new data points


```python
new_dataPoints = 5
x_new = np.linspace(0, 1, num=new_dataPoints)
t_new =  np.sin(2*np.pi*x_new)+0.2*np.random.normal(0,1,new_dataPoints)

y_pred = pn(x_new)
plt.plot(x_input, t_output, 'bo',x_input, yn, 'r-',x_new, y_pred, 'g--')
plt.legend(('Training data', 'Model', 'Testing data'))


err = 0.5*(np.sum((y_pred-t_new)**2))

print (err)#Testing error
```

# solving overfitting with higher number of dataset

```python
sample_point = 50

x_inpu = np.linspace(0,1, num = sample_point)

t_outpu = np.sin(2*np.pi*x_inpu) + 0.2*np.random.normal(0,1,sample_point)
Mv= 9
znew =  np.polyfit(x_inpu,t_outpu,Mv)
pnew = np.poly1d(znew)
ynew= pnew(x_inpu)


plt.plot(x_inpu, t_outpu, 'bo',x_inpu,ynew,'g-',x,sine,'r--')
plt.legend(('Training data', 'Model', 'sine wave'))


err = 0.5*(np.sum((ynew-t_outpu)**2))
print (err)#Testing error
```
